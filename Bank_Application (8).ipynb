```python
bank = {'acc':[1001,1002,1003,1004],
        'user':['Ravi','Raj','Simran','Rahul'],
        'pass':['Python@123','raj@123','simran@123','anjali@123'],
        'bal':[113.121,1231241.13412,124124124.213,1234134.343]}
```

```python
from getpass import getpass
print("*"*77)
print("-"*77)
print("ðŸ¤‘Welcome to Python BankðŸ¤‘".center(77))
print("-"*77)
print("*"*77)
admin = 'root'
passwd = 'redhat@123'
while True:
    print("\n$---Welcome To Main Menu---$")
    print("\n\t1. Login\n\t2. Signup\n\t3. Admin\n\t4. Exit")
    ch = input("\nEnter your choice --> ")
    if ch == "1":
        l_acc = int(input("\nEnter Account Number --> "))
        if l_acc in bank['acc']:
            index =bank['acc'].index(l_acc)
            l_pass = getpass("\nEnter your Account Password -->")
            if bank['pass'][index] == l_pass:
                while True:
                    print("\nWelcome to Login Menu!!")
                    print("\n\t1. Credit\n\t2. Debit\n\t3. Change Password\n\t4. Check Balance\n\t5. Logout")
                    l_ch = input("\nEnter your Choice --> ")
                    if l_ch == "1":
                        add_amt=float(input("enter amt to add : "))
                        bank['bal'][index]+=add_amt
                    elif l_ch == "2":
                        amt_with=float(input("enter amount to be withdrawn : "))
                        if(amt_with<=bank['bal'][index]):
                            bank['bal'][index]-=amt_with
                            print(f"{amt_with} is debited from your account .")
                        else:
                            print("insufficient amount !!")
     
                    elif l_ch == "3":
                        old_pass=getpass("enter yout old password : ")
                        if(old_pass==bank['pass'][index]):
                            l_new_pass = getpass("\nCreate a Password--> ")
                            l_v_pass = input("\nConfirm Your Password -->")
                            if l_new_pass == l_v_pass:
                                bank['pass'][index]=l_new_pass
                            else:
                                print("wrong password input!!")
                        else:
                            print("incorrect password !!!")
                    elif l_ch == "4":
                        print(f"\nYour Account Balance is {bank['bal'][index]}")
                    elif l_ch == "5":
                        l_ch1 = input("\nDo you really want to Exit?? (y/n) --> ").lower().strip()
                        if l_ch1 == "y" or l_ch1 == "yes":
                            print("\nYou are successfully Logged Out!!")
                            break
                        else:
                            continue
                    else:
                        print("\nInvalid Choice")
            else:
                print("\nPassword Doesnot Matched!!!")
        else:
            print("\nAccount doesnot Exists!!!")
    elif ch =="2":
        print("\nWelcome to SignUp Menu!!!")
        new_acc = bank['acc'][-1]+1
        new_bal = 0.0
        new_usr = input("\nEnter your UserName-->")
        new_pass = getpass("\nCreate a Password--> ")
        v_pass = input("\nConfirm Your Password -->")
        if new_pass == v_pass:
            bank['acc'].append(new_acc)
            bank['bal'].append(new_bal)
            bank['user'].append(new_usr)
            bank['pass'].append(new_pass)
            print(f"\nYour Account has been Created Successfully\nNow You can Login!!\nYour Account No. is  --> {new_acc}")
        else:
            print("\nPassword Doesnot Matched!!!")
    elif ch=="3":
        adm_usr = input("\nEnter Admin UserName --> ")
        if adm_usr == admin:
            adm_pass = getpass("\nEnter Admin Password --> ")
            if adm_pass == passwd:
                print("\nWelcome to Admin Panel!!!")
                s_acc = int(input("\nEnter an Account Number --> "))
                if s_acc in bank['acc']:
                    i =bank['acc'].index(s_acc)
                    print(f"\nUserName --> {bank['user'][i]}")
                    print(f"Password --> {bank['pass'][i]}")
                    print(f"Balance --> {bank['bal'][i]}")
                    print("\nThankYou!!")
                else:
                    print("\nAccount Number not found!!!")
            else:
                print("\nIncorrect Admin Password!!!")
        else:
            print("\nInvalid Admin User!!!")
    elif ch=="4":
        ch1 = input("\nDo you really want to Exit?? (y/n) --> ").lower().strip()
        if ch1 == "y" or ch1 == "yes":
            print("\nThankYou for Using Services!!")
            break
        else:
            continue
    else:
        print("\nInvalid Choice")
```
